```python
Add Vertices and Edges - add_vertex(v) - add_edge(u, v, weight=None)
Query - has_edge(u, v) - has_vertex(v)
```

```python
from dsa.graph import Graph

g = Graph.create_adjacency_matrix(directed=True, weighted=True)
```

```python
g.add_vertex("A")
g.add_vertex("B")
g.add_vertex("C")

g.print_graph()
```

```python
g.add_edge("A", "B", 1)
g.add_edge("A", "C", 2)
g.add_edge("B", "C", 3)

g.print_graph()
```

Query Methods

```python
g.vertices()
```

```python
g.edges()
```

```python
g.has_vertex("A") # True
g.has_vertex("B") # True
g.has_vertex("D") # False
```

```python
g.has_edge("A", "B") # True
g.has_edge("A", "C") # True
g.has_edge("C", "B") # False
```

```python
g.adjacents("A")
```

```python
g.adjacent_items("A")
```
